#### Check the correlation between each feature and the target (Scale).
- There are no features with high correlation with the target.
- All features will be kept in the dataset for processing because there is an indication that we can obtain better results if all features are used together to predict the target.

```python

df.drop(['recordNumber', 'Scale'], axis = 1).corrwith(df["Scale"]).abs().sort_values(ascending = False)
```

#### Check the Skewness of features.

```python

for i in df.drop(['recordNumber', 'Scale'], axis = 1).columns :
    print("Skewness for {} is {}".format(i, df[i].astype(float).skew()))
```

#### Check the Kurtosis of the features.

```python

for i in df.drop(['recordNumber', 'Scale'], axis = 1).columns :
    print("Kurtosis for {} is {}".format(i, df[i].astype(float).kurt()))
```

#### Check how the classes are balanced.
- The dataset is very unbalanced.
- There will be need to undersampling data for model training.

```python

instances_count = df['Scale'].value_counts().sum()
class_distribution = df['Scale'].value_counts().to_dict()
for key, value in class_distribution.items() :
    print(f"Class = {key:2.0f}   Qty = {value:6.0f}   Percentage = {(value / instances_count * 100):.2f} %")
```
